```python
import pandas as pd
import requests
import pathlib
import datetime as dt
```

```python
url = 'https://free-proxy-list.net/'
```

```python
rq = requests.get(url)
rq
```

```python
rq.content
```

```python
pd.read_html(rq.content)
```

```python
# Dataframe com lista de proxy disponível no site

df = pd.read_html(rq.content)[0]
df
```

```python
# Colunas do Dataframe
df.columns
```

```python
# Como os titulos de colunas devem ficar


[title.replace(' ','_').upper() for title in df.columns]
```

```python
# Atualização dos titulos de colunas do Dataframe

df.columns = [title.replace(' ','_').upper() for title in df.columns]
```

```python
# proxy que trabalham com proxy de https

df.HTTPS
```

```python
# Tratativa #1 em identificar proxy configurado como 'yes'

df.HTTPS.map(lambda x: x == 'yes')
```

```python
# Tratativa #2 em identificar proxy configurado como 'yes'

df.HTTPS.str.contains('y')
```

```python
# Filtragem dos proxies que trabalham com HTTPS

df.loc[df.HTTPS.str.contains('y')]
```

```python
# Criação de Dataset com subconjunto de dados filtrados

ds = df.loc[df.HTTPS.str.contains('y')]
ds
```

```python
help(ds.agg)
```

```python
# tratativa #1 mesclar colunas 

ds['IP_ADDRESS'] +":"+ ds["PORT"].astype(str)
```

```python
# tratativa #2 mesclar colunas 

ds[["IP_ADDRESS", "PORT"]].astype(str).apply(":".join, axis=1)
```

```python
# tratativa #3 mesclar colunas 


ds[['IP_ADDRESS', 'PORT']].astype(str).agg(':'.join, axis=1)
```

```python
# Campo com ip e porta mesclados

ds.loc[:, "PROXY"] = ds[['IP_ADDRESS', 'PORT']].astype(str).agg(':'.join, axis=1)
ds
```

```python
# Diretório de destino para arquivos produzidos

data_files = pathlib.Path().resolve().parents[2] / 'data_files'
```

```python
data_files.is_dir()
```

```python
# epoch com pacote datetime

dt.datetime.now().timestamp()
```

```python
# Isolar informação para gravar no arquivo CSV

ds[['PROXY']]
```

```python
# Gravando CSV com pandas

ds[['PROXY']].to_csv(data_files/f"proxies_{int(dt.datetime.now().timestamp())}.csv", index=False)
```

```python
# Gravando JSON com pandas

ds[['PROXY']].to_json(data_files/f"proxies_{int(dt.datetime.now().timestamp())}.json", orient='records', indent=4)
```
